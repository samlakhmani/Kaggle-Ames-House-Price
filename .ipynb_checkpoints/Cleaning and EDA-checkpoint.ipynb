```python
import pandas as pd
import numpy as np

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
test = pd.read_csv("./Data/test.csv")
train = pd.read_csv("./Data/train.csv")
```

```python
plot_data = train.select_dtypes(include='number')
```

```python
fig, axs = plt.subplots(nrows=(plot_data.shape[1]//3),ncols=3,figsize=(15,35))
fig.subplots_adjust(hspace = .5, wspace=.5)
axs = axs.ravel()

c=0
for i in plot_data.columns[:-1] :
    sns.regplot(x=i, y="SalePrice", data=plot_data, ax=axs[c])
    c+=1
```

```python
test['SalePrice'] = np.NaN
```

```python
comb = pd.concat ([train,test],axis=0)
```

```python
comb.set_index('Id',drop=True,inplace=True)
train.set_index('Id',drop=True,inplace=True)
test.set_index('Id',drop=True,inplace=True)
```

```python
fig, ax = plt.subplots(figsize=(20,20))
ax = sns.heatmap(
    comb.corr(), 
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(20, 220, n=200),
    square=True
    )
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
)
plt.show()
```

```python
corre=comb.corr()
corre = corre[(corre['SalePrice']>.5) | (corre['SalePrice']<-.1)]
corre['SalePrice']
```

```python
corre
```

```python
selected_features = corre.index
```

```python
corre = corre[selected_features]
```

```python
comb = comb[selected_features]
```

```python
for i in comb.columns[:-1]:
    print(comb[i].isnull().value_counts())
```
