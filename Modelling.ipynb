```python
import pandas as pd
import numpy as np

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train = pd.read_csv("train_clean.csv")
test = pd.read_csv("test_clean.csv")
comb = pd.concat([test,train],axis=0)

comb.set_index('Id',inplace=True)
```

```python
comb.corr()
```

```python
comb2 = pd.DataFrame()
for i in comb.columns:
    a = comb[i]
    if not i in ['YearBuilt','YearRemodAdd','TotalBsmtSF','1stFlrSF','GrLivArea','GarageArea','SalePrice']:
        a = pd.get_dummies(a)
    comb2 = pd.concat([comb2,a],axis=1)
comb2
```

```python
comb = comb2.sort_index(ascending=True)
```

```python
train = comb[comb['SalePrice'].notnull()]
test = comb[comb['SalePrice'].isnull()]
```

```python
test.drop('SalePrice',inplace=True,axis=1)
```

```python
X = train.drop('SalePrice',axis=1)
y = train['SalePrice']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=42)
```

### Random Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor

regr = RandomForestRegressor(max_depth=2, random_state=0)
regr.fit(X_train, y_train)
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(regr, X_train, y_train, cv=5)
scores.mean()
```

```python
regr.score(X_test,y_test)
```
